```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Load the Haar cascade classifier
classifier_face = cv2.CascadeClassifier("./haar_cascade/lib/python3.10/site-packages/cv2/data/haarcascade_frontalface_default.xml")

# Load the input image
image = cv2.imread('subject.jpg')

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect objects in the image using the classifier
objects_face = classifier_face.detectMultiScale(gray_image)

for (x, y, w, h) in objects_face:
    # cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)

    # Crop the object from the image
    get_face = image[y:y+h, x:x+w]

    # Save the cropped object to a file
    cv2.imwrite('cropped_object_face.jpg', get_face)
```

```python
face = cv2.imread('cropped_object_face.jpg')

plt.imshow(face)
```

```python
classifier_smile = cv2.CascadeClassifier("./haar_cascade/lib/python3.10/site-packages/cv2/data/haarcascade_smile.xml")

face_img = cv2.imread('cropped_object_face.jpg')

gray_face = cv2.cvtColor(face_img, cv2.COLOR_BGR2GRAY)

objects_smile = classifier_smile.detectMultiScale(gray_face)

for (x, y, w, h) in objects_smile:
    # cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)

    # Crop the object from the image
    get_smile = face_img[y:y+h, x:x+w]

    # Save the cropped object to a file
    cv2.imwrite('cropped_object_smile.jpg', get_smile)
```

```python
smile = cv2.imread('cropped_object_smile.jpg')

plt.imshow(smile)
```
